```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = {
    'Name':['Umer','Imtiaz','Hussna','Nazim','Hania','Malaika'],
    'Age':[19 , 21 , 18 , 17 , 20 , 22],
    'Spending':[100 , 200 , 300 , 110 , 290 , 130]
}
df = pd.DataFrame(data)
df
```

```python
X = df[['Age' , 'Spending']]
model = KMeans(n_clusters=2 , random_state=42 , n_init=10)

df['Group'] = model.fit_predict(X) 

plt.Figure(figsize=(6 , 4))
for group in df['Group'].unique():
    group_data = df[df['Group'] == group]
    plt.scatter(group_data['Age'] ,  group_data['Spending'] , label=f'Group {group}')

plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Customer Segments (K_Means)')
plt.legend()
plt.grid(True)
plt.show()

print(df)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
FruitBucket = {
    'Fruits':['Apple' , 'Orange' , 'Banana' , 'Cherry'],
    'Colors':['Red' , 'OrangeColor' , 'Yellow' , 'Red'],
    'Size(cm)':[32.99 , 19.11 , 11.21 , 5.11],
    'Quantity':[100 , 50 , 77 , 19]
    }

df = pd.DataFrame(FruitBucket)
df
```

```python
X = df[['Size(cm)' , 'Quantity']]
model = KMeans(n_clusters=4 , random_state=42 , n_init=10)

df['Group'] = model.fit_predict(X)

plt.Figure(figsize=(6 , 4))
for group in df['Group'].unique():
    group_data = df[df['Group'] == group]
    plt.scatter(group_data['Size(cm)'] , group_data['Quantity'] , label= f'group {group}') 
plt.xlabel('Size(cm)')
plt.ylabel('Quantity')
plt.title('Fruit Clusters (Based On Pattern (k-Means))')
plt.legend()
plt.grid(True)
plt.show()

print(df)    
```
